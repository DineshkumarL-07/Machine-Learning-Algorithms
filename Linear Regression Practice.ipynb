Linear Regression Practice

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\rgowt\Downloads\canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('capita income')
plt.scatter(df.year, df.per_capita_income,color='blue')
```

```python
reg = LinearRegression()
```

```python
reg.fit(df[['year']],df.per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=15)
plt.ylabel('capita income',fontsize=15)
plt.scatter(df.year,df.per_capita_income,color='red')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
